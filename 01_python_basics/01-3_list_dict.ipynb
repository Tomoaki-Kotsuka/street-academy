## リストとは

リストは **順番に並んだ複数の値をまとめて持つ** ための型です。「売上のリスト」「顧客名のリスト」のように、Excel の列 1 つ分のデータを表現するのに向いています。

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 作る（リテラル） | `Array(1, 2, 3)` | `[1, 2, 3]` |
| 動的な配列 | `Dim arr()` + `ReDim Preserve` | `arr = []`（最初から動的）|
| 要素アクセス | `arr(0)` | `arr[0]`（どちらも 0 始まり）|
| 長さ | `UBound(arr) + 1` | `len(arr)` |
| 末尾に追加 | `ReDim Preserve` + 代入 | `arr.append(x)` |

Python のリストは **後から要素を追加・削除できる動的な配列** で、VBA の `Collection` に近い感覚です。

## やってみる: リストを作って中身を見る

````python
# リストを作る
sales = [1980, 2200, 1500, 3100, 980]

# 全部表示
print(sales)

# 個別アクセス（0 始まり）
print(sales[0])    # → 1980（先頭）
print(sales[-1])   # → 980（末尾）

# スライス（[開始:終了]、終了は含まない。1-2 の文字列スライスと同じ記法）
print(sales[1:3])  # → [2200, 1500]

# 長さ
print(len(sales))  # → 5

# 含まれているか
print(1980 in sales)  # → True
print(9999 in sales)  # → False
````

## リストの追加・削除・更新

### よく使うメソッド

| やること | コード | VBA で言うと |
|---|---|---|
| 末尾に追加 | `lst.append(x)` | `coll.Add x`（Collection）|
| 末尾を取り出す | `lst.pop()` | （標準なし）|
| 値を指定して削除 | `lst.remove(x)` | `coll.Remove(idx)` 相当 |
| インデックスで削除 | `del lst[i]` | `coll.Remove(idx)` |
| 連結 | `lst1 + lst2` | 複数の Add ループが必要 |

````python
items = ["りんご", "ばなな"]

# 末尾に追加
items.append("みかん")
print(items)   # → ['りんご', 'ばなな', 'みかん']

# 値で削除（最初に見つかったものが消える）
items.remove("ばなな")
print(items)   # → ['りんご', 'みかん']

# インデックスで削除
del items[0]
print(items)   # → ['みかん']

# 更新（直接代入）
items.append("ぶどう")
items[0] = "メロン"
print(items)   # → ['メロン', 'ぶどう']

# リスト同士の連結
fruits = ["りんご", "ばなな"]
veggies = ["きゅうり", "トマト"]
all_items = fruits + veggies
print(all_items)   # → ['りんご', 'ばなな', 'きゅうり', 'トマト']
````

## 並び替え・合計・最大最小

リストに対しては、組み込み関数で **集計や並び替えがすぐにできます**。Excel の `SUM`, `MAX`, `MIN`, `AVERAGE` をプログラムから直接呼ぶ感覚です。

````python
sales = [1980, 2200, 1500, 3100, 980]

# 合計
print(sum(sales))              # → 9760

# 最大・最小
print(max(sales))              # → 3100
print(min(sales))              # → 980

# 平均（合計 / 件数）
print(sum(sales) / len(sales)) # → 1952.0

# 並び替え（新しいリストを返す。元は変わらない）
print(sorted(sales))                # → [980, 1500, 1980, 2200, 3100]
print(sorted(sales, reverse=True))  # → [3100, 2200, 1980, 1500, 980]

# 並び替え（元のリスト自体を変える）
sales.sort()
print(sales)   # → [980, 1500, 1980, 2200, 3100]
````

## リストの中身を1つずつ取り出す（おまけ）

`for` ループを使うと、リストの中身を順番に取り出して処理できます。`for` 自体は次のノートブック 1-4 でじっくり扱いますが、**「こういう書き方ができる」だけ** 見ておきます。

````python
items = ["りんご", "ばなな", "みかん"]

for item in items:
    print(f"商品: {item}")

# → 商品: りんご
# → 商品: ばなな
# → 商品: みかん
````

## 辞書とは

辞書は **「キー（名前）」と「値」のペアをまとめて持つ** ための型です。「商品名 → 単価」「社員番号 → 氏名」のような **対応関係** を表現するのに使います。

Excel で言うと **VLOOKUP の対応表** に近いイメージです。

### VBA との対比

| やること | VBA (Scripting.Dictionary) | Python |
|---|---|---|
| 作る | `Set d = CreateObject("Scripting.Dictionary")`<br>`d.Add "name", "田中"` | `d = {"name": "田中"}` |
| 値を取得 | `d.Item("name")` | `d["name"]` |
| 追加・更新 | `d.Add` / `d.Item("k") = x` | `d["k"] = x` |
| キーが存在するか | `d.Exists("k")` | `"k" in d` |
| 削除 | `d.Remove("k")` | `del d["k"]` |
| 全キー | `d.Keys` | `d.keys()` |

Python の辞書は **言語に標準で組み込まれている** ので、VBA のように外部参照（Microsoft Scripting Runtime）を追加する必要がありません。

````python
# 辞書を作る（キー: 値 の形）
person = {
    "name": "田中",
    "age": 30,
    "department": "営業部",
}

print(person)

# キーで値を取り出す
print(person["name"])         # → 田中
print(person["age"])          # → 30

# キーが存在するか
print("name" in person)        # → True
print("salary" in person)      # → False

# 長さ（キーの数）
print(len(person))             # → 3
````

## 辞書の追加・更新・削除

````python
person = {"name": "田中", "age": 30}

# 追加（新しいキーに代入）
person["department"] = "営業部"
print(person)
# → {'name': '田中', 'age': 30, 'department': '営業部'}

# 更新（既存のキーに代入）
person["age"] = 31
print(person)
# → {'name': '田中', 'age': 31, 'department': '営業部'}

# 削除
del person["department"]
print(person)
# → {'name': '田中', 'age': 31}
````

## キー一覧・値一覧・キーと値のペア

辞書全体を巡回したいときに使います。

````python
prices = {
    "りんご": 150,
    "ばなな": 100,
    "みかん": 80,
}

# キー一覧
print(list(prices.keys()))     # → ['りんご', 'ばなな', 'みかん']

# 値一覧
print(list(prices.values()))   # → [150, 100, 80]

# キーと値のペアを取り出す（for と組み合わせるのが定番）
for name, price in prices.items():
    print(f"{name}: ¥{price}")

# → りんご: ¥150
# → ばなな: ¥100
# → みかん: ¥80

# 値の合計
print(sum(prices.values()))    # → 330
````

## ちょっと実務寄り: リストと辞書を組み合わせる

「**複数の人 / 複数の商品データをまとめて持つ**」ときは、**辞書のリスト** で表現するのが定番です。これは次章で扱う pandas の DataFrame の考え方そのものに繋がります。

````python
# 社員データ（辞書のリスト = 1 人 1 辞書 を並べる）
employees = [
    {"id": 1, "name": "田中", "department": "営業", "salary": 350000},
    {"id": 2, "name": "鈴木", "department": "開発", "salary": 420000},
    {"id": 3, "name": "佐藤", "department": "営業", "salary": 380000},
]

# 1人目の情報
print(employees[0])
# → {'id': 1, 'name': '田中', 'department': '営業', 'salary': 350000}

# 1人目の名前だけ
print(employees[0]["name"])    # → 田中

# 全員を順に表示
for e in employees:
    print(f"{e['name']} ({e['department']}) ¥{e['salary']:,}")
````

## よくあるエラーと対処

### `IndexError: list index out of range`

```python
items = [1, 2, 3]
print(items[5])  # IndexError
```

→ リストの範囲外を指定している。`len(items)` で長さを確認してからアクセスする。

### `KeyError: 'xxx'`

```python
d = {"a": 1}
print(d["b"])  # KeyError
```

→ 辞書にないキーを指定している。`"b" in d` で存在チェックするか、`d.get("b", デフォルト値)` を使う：

```python
print(d.get("b", 0))   # → 0（キーがない時のデフォルト）
```

### `AttributeError: 'list' object has no attribute 'add'`

```python
items = []
items.add(1)   # ❌ そんなメソッドはない
```

→ リストに要素を追加するのは `append()`。`add()` ではないので注意（VBA の Collection の `Add` と混同しがち）。

## まとめ

このノートブックで学んだこと：

- **リスト** `[...]` は **順番に並んだ値の集まり**
  - 追加 `.append()`、削除 `del` / `.remove()` / `.pop()`
  - 集計 `sum()` `max()` `min()` `len()` を直接使える
  - 並び替え `sorted()`（新しいリスト返却）/ `.sort()`（元を変える）
  - 含まれているか: `in`
- **辞書** `{...}` は **キーと値のペアの集まり**
  - アクセス `d[key]`、追加・更新 `d[key] = ...`、削除 `del d[key]`
  - 巡回 `.keys()` / `.values()` / `.items()`
  - キー存在チェック: `in`、キー欠落時のデフォルト: `d.get(k, default)`
- 「**複数のレコードをまとめて持つ**」とき → **辞書のリスト** が定番（→ 次章 pandas の DataFrame に直結する考え方）
- **リスト内包表記** `[式 for x in リスト]` で `for + append` を 1 行に
