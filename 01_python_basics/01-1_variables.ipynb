## 変数とは

変数は **値に名前を付けて取っておく箱** です。VBA で書いていた `Dim x As Integer` と同じ役割を果たします。

ただし、Python では **型を宣言する必要がありません**。値を代入するだけで変数ができあがります。

### VBA との対比

| | VBA | Python |
|---|---|---|
| 整数 | `Dim age As Integer`<br>`age = 30` | `age = 30` |
| 小数 | `Dim price As Double`<br>`price = 1980.5` | `price = 1980.5` |
| 文字列 | `Dim name As String`<br>`name = "田中"` | `name = "田中"` |
| 真偽値 | `Dim active As Boolean`<br>`active = True` | `active = True` |

Python は **値から型を自動で判断** するので、`Dim ... As ...` を書きません。短く書ける反面、「何が入っているか後から見て分かりにくい」というデメリットもあります（そこは後で `type()` で補えます）。

## やってみる: 変数を作って表示する

下のセルを **Shift + Enter** で実行してみてください。

````python
# 数値（整数）
age = 30

# 数値（小数）
price = 1980.5

# 文字列（ダブルクォート " でもシングルクォート ' でも OK）
name = "田中"

# 真偽値（先頭の T / F は大文字。VBA と同じ）
active = True

# print() で中身を表示
print(age)
print(price)
print(name)
print(active)
````

## 型を確認する

VBA だと宣言時に `As Integer` で型が分かりましたが、Python では `type()` という関数で確認します。

````python
print(type(age))
print(type(price))
print(type(name))
print(type(active))
````

出力された型の名前と、よく使う対応表：

| Python の型 | 意味 | VBA で言うと |
|---|---|---|
| `int` | 整数 | `Integer` / `Long` |
| `float` | 小数 | `Double` / `Single` |
| `str` | 文字列 | `String` |
| `bool` | 真偽値 (True / False) | `Boolean` |

## 計算してみる

変数同士で計算ができます。基本の四則演算は VBA と似ていますが、いくつか書き方が違います。

````python
# 単価と数量から金額を計算
price = 1980
quantity = 3

print(price + quantity)    # 足し算 → 1983
print(price - quantity)    # 引き算 → 1977
print(price * quantity)    # 掛け算（単価 × 数量）→ 5940（合計金額）

# 割り勘の計算
total = 10000
people = 3

print(total / people)      # 割り算（小数あり） → 3333.333...
print(total // people)     # 整数の割り算（切り捨て） → 3333
print(total % people)      # 余り → 1

# べき乗（2の10乗）
print(2 ** 10)             # → 1024
````

### ちなみに、文字列にも `+` が使えます

数値の `+` は足し算でしたが、文字列に対しては **連結** になります。VBA でいう `&` と同じ役割です。

````python
# 文字列の連結
first_name = "田中"
last_name = "太郎"

print(first_name + last_name)         # → 田中太郎
print(first_name + " " + last_name)   # → 田中 太郎

# おまけ: `*` は文字列に使うと「繰り返し」になります
print("-" * 30)                       # → ------------------------------
````

### VBA との演算子の違いに注意

| 演算 | VBA | Python |
|---|---|---|
| 割り算（小数あり） | `/` | `/` |
| 整数の割り算（切り捨て） | `\` | `//` |
| 余り | `Mod` | `%` |
| べき乗 | `^` | `**` |

特に **`^` は Python では別の意味になる** ので、`5の3乗` を書きたいときは `5 ** 3` です。

## 型を変換する

文字列と数値を混ぜて扱おうとすると、エラーになることがあります。例えば、フォームから入力された「金額」は文字列で受け取られるので、計算するには数値に変換する必要があります。

### 主な変換関数

| 関数 | 何をする | VBA で言うと |
|---|---|---|
| `int("123")` | 文字列 → 整数 | `CInt()` |
| `float("1.5")` | 文字列 → 小数 | `CDbl()` |
| `str(123)` | 数値 → 文字列 | `CStr()` |

````python
# 文字列 → 数値に変換してから計算
price_text = "1980"
tax = 198

total = int(price_text) + tax
print(total)  # → 2178
````

````python
# 数値 → 文字列に変換してから連結
age = 30
message = "私は " + str(age) + " 歳です"
print(message)  # → 私は 30 歳です
````

## 変数の命名ルール

Python での変数名のルール：

- **使える文字**: 半角英数字とアンダースコア `_`（日本語も使えますが推奨されません）
- **数字で始めるのは NG**: `1st_name` ❌ → `first_name` ⭕
- **大文字小文字を区別する**: `age` と `Age` は別の変数（**VBA は区別しないので、ここは大きな違い**）
- **慣習: スネークケース**: 単語の区切りは `_` を使う（`firstName` ではなく `first_name`）
- **予約語は使えない**: `if`, `for`, `class`, `True` などは変数名にできない

### 良い例 / 悪い例

```
⭕ first_name = "田中"
⭕ total_price = 1980
⭕ is_active = True

❌ 1st_name = "田中"      # 数字で始まっている
❌ class = "営業部"        # class は予約語
❌ first-name = "田中"     # ハイフンは使えない（引き算と区別がつかない）
```

## よくあるエラーと対処

### `NameError: name 'xxx' is not defined`

→ 変数がまだ定義されていないか、スペルミス。
特に **VBA と違って大文字小文字を区別** するので注意。

```python
name = "田中"
print(Name)  # NameError: name 'Name' is not defined
```

### `TypeError: can only concatenate str (not "int") to str`

→ 文字列と数値を直接 `+` で繋ごうとした。`str()` で変換するか、f-string を使う。

```python
age = 30
print("年齢: " + age)        # ❌ TypeError
print("年齢: " + str(age))   # ⭕ 文字列に変換すれば OK
```

### `SyntaxError: invalid syntax`

→ 構文の書き間違い。よくあるのは **`=`（代入）と `==`（等価比較）の混同** や、**括弧・クォートの閉じ忘れ**。
