## 条件分岐 `if` / `else`

「値によって処理を変える」ための構文です。VBA の `If ... Then ... Else ... End If` と同じ役割ですが、Python は **インデント（字下げ）で処理のブロックを表現する** のが大きな違いです。

### VBA との対比

```vba
' VBA
If score >= 80 Then
    result = "合格"
Else
    result = "不合格"
End If
```

```python
# Python
if score >= 80:
    result = "合格"
else:
    result = "不合格"
```

- 行末は `:`（コロン）
- ブロックの中身は **半角スペース4つ** で字下げ（Colab では自動で入ります）
- `End If` のような閉じ記号は **不要**（インデントが戻った時点でブロック終了）

````python
score = 85

if score >= 80:
    print("合格")
else:
    print("不合格")

# 値を変えて何度か試してみてください
score = 60
if score >= 80:
    print("合格")
else:
    print("不合格")
````

## 3つ以上に分ける `elif`

条件が3つ以上ある場合は `elif`（else if の略）でつなげます。VBA の `ElseIf` と同じです。

上から順番に評価され、**最初に True になった枝だけが実行されます**。

````python
score = 75

if score >= 90:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 50:
    grade = "C"
else:
    grade = "D"

print(f"スコア {score} → 評価 {grade}")  # → スコア 75 → 評価 B
````

## 比較演算子と論理演算子

### 比較演算子（結果は `True` / `False` の bool）

| 意味 | Python | VBA |
|---|---|---|
| 等しい | `==` | `=` |
| 等しくない | `!=` | `<>` |
| より大きい | `>` | `>` |
| 以上 | `>=` | `>=` |
| より小さい | `<` | `<` |
| 以下 | `<=` | `<=` |

> ⚠️ Python では **代入は `=`、比較は `==`**。VBA は両方 `=` で文脈で判断するので、ここは特に混同しやすい場所です。

### 論理演算子（複数の条件を組み合わせる）

| 意味 | Python | VBA |
|---|---|---|
| かつ | `and` | `And` |
| または | `or` | `Or` |
| 否定 | `not` | `Not` |

### `in` で「含まれているか」を判定

`x in リスト` でリストに含まれるか、`key in 辞書` で辞書のキーに含まれるかを bool で返します（1-3 で出てきた `in` の本来の使い方）。

````python
# 比較は True / False を返す
print(10 == 10)   # → True
print(10 != 5)    # → True
print(10 < 5)     # → False

# 論理演算子で複数条件を組み合わせる
age = 35
department = "営業"

if age >= 30 and department == "営業":
    print("30歳以上の営業の方です")

if age < 20 or age >= 60:
    print("対象外です")
else:
    print("対象です")

# in で含まれているか判定
vip_customers = ["田中商事", "鈴木株式会社"]
customer = "田中商事"

if customer in vip_customers:
    print(f"{customer} は VIP です")

# 否定
if "佐藤製作所" not in vip_customers:
    print("佐藤製作所 は VIP ではありません")
````

## ループ `for`

`for` は **「リストや範囲の中身を順番に取り出して、同じ処理を繰り返す」** ための構文です。pandas で行を処理したり、複数ファイルを一括処理したりするときの土台になります。

### VBA との対比

```vba
' VBA: 回数指定
For i = 1 To 5
    Cells(i, 1).Value = i
Next i

' VBA: コレクションを順に
For Each item In items
    Debug.Print item
Next item
```

```python
# Python: 回数指定（range で 0〜4 が出る）
for i in range(5):
    print(i)

# Python: リスト/辞書を順に（VBA の For Each に相当）
for item in items:
    print(item)
```

Python の `for` は **常に「何かを順番に取り出す」形** です。VBA の `For i = 1 To 5` のような回数指定がしたいときは、`range()` を組み合わせます。

### `range()` の3パターン

| 書き方 | 出てくる値 | VBA で言うと |
|---|---|---|
| `range(5)` | `0, 1, 2, 3, 4` | `For i = 0 To 4` |
| `range(1, 6)` | `1, 2, 3, 4, 5` | `For i = 1 To 5` |
| `range(0, 10, 2)` | `0, 2, 4, 6, 8` | `For i = 0 To 8 Step 2` |

**終了の値は含まない**（1-2 のスライスと同じルール）のがポイントです。

````python
# 0 から 4 まで
for i in range(5):
    print(i)

print("---")

# 1 から 5 まで（終了の 6 は含まない）
for i in range(1, 6):
    print(f"{i}回目")

print("---")

# 0 から 10 未満まで 2 ずつ
for i in range(0, 10, 2):
    print(i)   # → 0, 2, 4, 6, 8
````

### リスト・辞書を `for` で回す

1-3 でチラ見せした書き方を、ここでしっかり押さえます。

````python
# リストを順に
items = ["りんご", "ばなな", "みかん"]
for item in items:
    print(f"商品: {item}")

print("---")

# インデックスもほしい時は enumerate
for i, item in enumerate(items, start=1):
    print(f"{i}番目: {item}")

print("---")

# 辞書はキーと値を items() で同時に取り出すのが定番
prices = {"りんご": 150, "ばなな": 100, "みかん": 80}
for name, price in prices.items():
    print(f"{name}: ¥{price}")
````

## ループの中で集計する

実務でとてもよく出てくる **「ループしながら合計を出す」「条件に合うものだけ数える」** パターンです。

VBA でよく書く `total = total + Cells(i, 1).Value` の Python 版で、コツは **`for` の前に変数を初期化しておく** ことです。

````python
sales = [12000, 8500, 15000, 9800, 22000, 7600]

# 合計を計算
total = 0
for s in sales:
    total = total + s   # 「total += s」とも書ける（VBA にはない短縮形）
print(f"合計: ¥{total:,}")

# 1万円以上の件数を数える
count = 0
for s in sales:
    if s >= 10000:
        count += 1
print(f"1万円以上: {count}件")

# もちろん sum() / len() を使えば 1 行で書ける
print(f"合計（短縮版）: ¥{sum(sales):,}")
print(f"件数: {len(sales)}件")
````

## ループ `while`

`while` は **条件が True の間ずっと繰り返す** ループです。回数が決まっていない処理（「閾値を超えるまで」「ユーザーが終了と入力するまで」等）に使います。

VBA の `Do While ... Loop` と同じ感覚です。

> ⚠️ 条件が永久に True のままだと **無限ループ** になります。Colab は右上の停止ボタンで止められますが、必ず **「ループの中で条件が変化する」ように書く** のがお約束です。

````python
# 1, 2, 4, 8, ... と倍々にして、100 を超えたら止める
n = 1
while n < 100:
    print(n)
    n = n * 2   # ← この更新を忘れると無限ループ

print(f"100 を超えた値: {n}")
````

## ループを途中で抜ける / スキップする

- `break` — ループ自体を抜ける（残りを処理しない）
- `continue` — 今回の処理だけスキップして、次の繰り返しへ進む

VBA の `Exit For` / `Exit Do` が `break` に相当します（`continue` は VBA に直接の相当物がなく、`If ... Then ... End If` で囲って実現します）。

````python
# break: 最初に見つかった大口取引で抜ける
sales = [3000, 5000, 12000, 800, 25000]
for s in sales:
    if s >= 10000:
        print(f"大口取引を発見: ¥{s:,}")
        break
    print(f"確認中: ¥{s:,}")

print("---")

# continue: マイナス値（返品）はスキップして集計
amounts = [1000, -500, 2000, -300, 1500]
total = 0
for a in amounts:
    if a < 0:
        continue   # 以降の処理をスキップして次の a へ
    total += a
print(f"プラス分のみ合計: ¥{total:,}")   # → 4500
````

## リスト内包表記（本格導入）

1-3 の最後にチラ見せした書き方をしっかり扱います。**`for + append` を 1 行で書ける** Python 独特の構文で、データ処理では頻出します。

### 基本形

```python
# for ループ版
results = []
for x in リスト:
    results.append(式)

# リスト内包表記版（同じ意味）
results = [式 for x in リスト]
```

### 条件付き

```python
# for + if 版
results = []
for x in リスト:
    if 条件:
        results.append(式)

# リスト内包表記版
results = [式 for x in リスト if 条件]
```

**読み方のコツ**: 後ろから `for` と `if` を読むと、頭の中で for ループに展開できます。

````python
sales = [12000, 8500, 15000, 9800, 22000, 7600]

# 全部を税込み（1.1倍）にする
tax_included = [s * 1.1 for s in sales]
print(tax_included)

# 1万円以上だけ抽出
big = [s for s in sales if s >= 10000]
print(big)   # → [12000, 15000, 22000]

# 顧客データから「VIP かつ売上1万以上」の名前だけ抽出
customers = [
    {"name": "田中商事", "amount": 80000, "vip": True},
    {"name": "鈴木株式会社", "amount": 5000, "vip": True},
    {"name": "佐藤製作所", "amount": 50000, "vip": False},
    {"name": "山田工業", "amount": 30000, "vip": True},
]
vip_names = [c["name"] for c in customers if c["vip"] and c["amount"] >= 10000]
print(vip_names)   # → ['田中商事', '山田工業']
````

## ちょっと実務寄り: 条件分岐 + ループの組み合わせ

**売上リストを集計して、ランクごとに件数と合計を出す** — こういう処理は実務でも毎日のように書きます。`for` + `if/elif/else` + 辞書 を組み合わせる、これまでの総復習になる例です。

````python
# 1日分の売上データ
sales = [12000, 8500, 15000, 9800, 22000, 7600, 35000, 4200, 18000]

# ランク別の集計箱を辞書で用意
summary = {
    "大口 (2万以上)": {"count": 0, "total": 0},
    "中口 (1万〜2万未満)": {"count": 0, "total": 0},
    "小口 (1万未満)": {"count": 0, "total": 0},
}

for s in sales:
    if s >= 20000:
        rank = "大口 (2万以上)"
    elif s >= 10000:
        rank = "中口 (1万〜2万未満)"
    else:
        rank = "小口 (1万未満)"

    summary[rank]["count"] += 1
    summary[rank]["total"] += s

# 結果を表示
for rank, data in summary.items():
    print(f"{rank:<20} {data['count']}件   合計 ¥{data['total']:,}")
````

## よくあるエラーと対処

### `IndentationError: expected an indented block`

```python
if score >= 80:
print("合格")   # ❌ インデントしていない
```

→ `if` / `for` / `while` などのブロックは **必ずインデント** が必要。Colab では Enter を押すと自動でインデントが入ります。

### `IndentationError: unexpected indent` / `TabError`

→ インデント幅がバラバラ（半角スペース2と4が混在、タブとスペースが混在 等）。Python では **同じブロックの中はインデント幅を揃える** ことが必須です。Colab で書く分には自動的に揃うので心配ありませんが、コピペした行は気をつけてください。

### `SyntaxError: invalid syntax`（`if x = 10:` で発生）

```python
if score = 80:   # ❌ 代入と比較を混同
    print("ぴったり")
```

→ Python では **代入は `=`、比較は `==`**。VBA は両方 `=` なので、ここはよく間違える場所です。

### `if` の条件のコロン `:` 忘れ

```python
if score >= 80   # ❌ 行末の : が抜けている
    print("合格")
```

→ `if` / `elif` / `else` / `for` / `while` の行末は **必ず `:`**。Colab では赤い波線で教えてくれます。

### 無限ループしてしまった

→ Colab の左側、実行中セルの **停止ボタン（■）** で止められます。`while` の中で **条件に使っている変数を必ず更新する** ように書き直してください。

## まとめ

このノートブックで学んだこと：

- **条件分岐** `if` / `elif` / `else` で値によって処理を分ける
  - 行末は `:`、ブロックは **インデント** で表す（`End If` 不要）
  - 代入 `=` と比較 `==` は別物
- **比較演算子** `== != < > <= >=`、**論理演算子** `and` / `or` / `not`、**含有チェック** `in`

- **`break`** でループ脱出、**`continue`** で今回だけスキップ
- **リスト内包表記** `[式 for x in リスト if 条件]` — 絞り込み + 加工を1行で
- 集計の定番パターン: **辞書を集計箱にして `for` で回す**（次章 pandas の `groupby` の原型）
