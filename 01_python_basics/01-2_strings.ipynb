## 文字列の作り方

文字列はダブルクォート `"..."` でも、シングルクォート `'...'` でも作れます。両者に機能的な違いはありません。

```python
name1 = "田中"
name2 = '田中'
```

ただし、文字列の中にクォートを含めたい場合は使い分けると便利です。

- `"It's a test"` ← 中にシングル `'` が含まれるならダブルで囲む
- `'She said "hello"'` ← 中にダブル `"` が含まれるならシングルで囲む

### エスケープ

同じ種類のクォートを中に書きたい、改行やタブを入れたい場合は、バックスラッシュ `\` でエスケープします。

| 書き方 | 意味 |
|---|---|
| `\"` | ダブルクォート |
| `\'` | シングルクォート |
| `\n` | 改行 |
| `\t` | タブ |
| `\\` | バックスラッシュそのもの |

````python
# クォートの使い分け
print("It's a test")
print('She said "hello"')

# エスケープ
print("She said \"hello\"")
print("1行目\n2行目")
print("名前\t年齢")
````

## 連結と長さ

連結は `+`（1-1 でやりました）。文字列の長さは `len()` で取れます。

### VBA との対比

| | VBA | Python |
|---|---|---|
| 連結 | `"Hello " & "World"` | `"Hello " + "World"` |
| 長さ | `Len(s)` | `len(s)` |
| 繰り返し | `String(5, "-")` | `"-" * 5` |

````python
greeting = "こんにちは"
name = "田中さん"

# 連結
message = greeting + "、" + name
print(message)

# 長さ（文字数）
print(len(message))

# 繰り返し
print("=" * 30)
````

## f-string で変数を埋め込む

`+` 連結は単純ですが、変数が増えると `+` だらけで読みにくく、しかも数値は `str()` で変換する手間も発生します。

そこで Python 3.6 以降の標準として使われているのが **f-string（エフストリング）** です。

### 書き方は2つだけ

1. 文字列の前に `f` を付ける（**忘れるとただの文字列扱い** になります）
2. 中に埋め込みたい変数を `{ }` で囲む

````python
name = "田中"
age = 35

# + 連結だと型変換 (str) が必要で、繋ぎ目もごちゃつく
print("私は " + name + " です。" + str(age) + " 歳です。")

# f-string ならスッキリ
print(f"私は {name} です。{age} 歳です。")

# 中で計算もできる
print(f"来年は {age + 1} 歳になります。")
````

## f-string の書式指定（ここからが本領発揮）

`{変数:書式}` の `書式` 部分を使うと、Excel の「セルの書式設定」のような出力が **コード1行** で作れます。事務職の業務報告フォーマットを作るときの強力な武器です。

### よく使う書式指定

| 書式 | 効果 | 例 | 結果 |
|---|---|---|---|
| `:,` | 3桁区切り | `f"{1980000:,}"` | `1,980,000` |
| `:.2f` | 小数点以下2桁 | `f"{3.14159:.2f}"` | `3.14` |
| `:.1%` | パーセント表示 | `f"{0.157:.1%}"` | `15.7%` |
| `:03d` | 整数のゼロ埋め（3桁） | `f"{7:03d}"` | `007` |
| `:<10` | 左寄せ（幅10） | `f"\|{'A':<10}\|"` | `\|A         \|` |
| `:>10` | 右寄せ（幅10） | `f"\|{'A':>10}\|"` | `\|         A\|` |

組み合わせも可能です。例: `:>10,` は「右寄せ幅10 + 3桁区切り」。

````python
# 売上金額を3桁区切りで
price = 1980000
print(f"売上: ¥{price:,}")

# 消費税率をパーセント表示
tax_rate = 0.1
print(f"消費税率: {tax_rate:.1%}")

# 平均値を小数点以下2桁で
average = 87.5294
print(f"平均点: {average:.2f}")

# 月をゼロ埋め2桁
month = 3
print(f"2025年{month:02d}月のレポート")

# 数字を桁揃え（右寄せ8桁 + 3桁区切り）
print(f"|{1980000:>10,}|")
print(f"|{   1980:>10,}|")
print(f"|{    120:>10,}|")
````

> 💡 **日本語の桁揃えに関する注意**: `:<10` の `10` は「**文字数** 10」なので、日本語など全角文字が混じると見た目（半角換算で2文字分の幅）が完全には揃いません。整形した表が必要なときは、英数字部分だけ揃える、全角スペースで詰める、などの工夫が必要になります。

## 大文字小文字の変換

データクレンジング（表記揺れ整理）でよく使います。例えばメールアドレスを全て小文字に統一して重複判定する、など。

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 大文字に | `UCase(s)` | `s.upper()` |
| 小文字に | `LCase(s)` | `s.lower()` |
| 先頭だけ大文字 | （標準なし） | `s.capitalize()` |
| 各単語の先頭を大文字 | `StrConv(s, vbProperCase)` | `s.title()` |

````python
email = "TANAKA@Example.com"

print(email.upper())       # → TANAKA@EXAMPLE.COM
print(email.lower())       # → tanaka@example.com
print(email.capitalize())  # → Tanaka@example.com

name = "yamada taro"
print(name.title())        # → Yamada Taro
````

## 検索と置換

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 含まれているか | `InStr(s, "x") > 0` | `"x" in s` |
| 位置を取得 | `InStr(s, "x")` | `s.find("x")` |
| 置換 | `Replace(s, old, new)` | `s.replace(old, new)` |
| 先頭判定 | （複数関数の組合せ） | `s.startswith("...")` |
| 末尾判定 | （複数関数の組合せ） | `s.endswith("...")` |

`in` 演算子は **「含まれているか」だけを True/False で返す** ので、Excel の `ISNUMBER(SEARCH(...))` のような書き方より直感的です。

````python
text = "今日の売上は1980円でした"

# 含まれているか（True / False が返る）
print("売上" in text)         # → True
print("仕入れ" in text)       # → False

# 置換（元の text は変わらず、新しい文字列が返る）
print(text.replace("1980", "2200"))  # → 今日の売上は2200円でした

# 先頭・末尾の判定（ファイル名チェックで便利）
filename = "report_2025-03.xlsx"
print(filename.endswith(".xlsx"))     # → True
print(filename.startswith("report"))  # → True
````

## 分割と結合

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 区切って分割 | `Split(s, ",")` | `s.split(",")` |
| リストを連結 | `Join(arr, ",")` | `",".join(list)` |

`split()` の結果は **リスト**（複数の値をまとめて持つもの）になります。リスト自体は次のノートブック `01-3_list_dict.ipynb` で詳しく扱いますが、ここでは「`[0]` で1つ目、`[1]` で2つ目を取り出せる」とだけ覚えてください。

````python
# CSV風の文字列を区切って分割
csv_line = "田中,営業部,35"
parts = csv_line.split(",")
print(parts)        # → ['田中', '営業部', '35']

# 個別に取り出す（0始まり）
print(parts[0])     # → 田中
print(parts[1])     # → 営業部
print(parts[2])     # → 35

# リストを連結（区切り文字を間に挟む）
words = ["こんにちは", "Python", "の世界"]
print(" / ".join(words))  # → こんにちは / Python / の世界
````

## 前後の空白を除去 (strip)

Excel からコピペしたデータには、前後に余計な空白が混じることがよくあります。事務処理では出番が多い操作です。

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 前後の空白除去 | `Trim(s)` | `s.strip()` |
| 先頭の空白のみ | `LTrim(s)` | `s.lstrip()` |
| 末尾の空白のみ | `RTrim(s)` | `s.rstrip()` |

`strip()` は引数に文字を渡すと、その文字を除去できます（例: `s.strip("*")` で `*` を除去）。

````python
raw = "   田中 太郎   "
print(f"[{raw.strip()}]")     # → [田中 太郎]

# 改行も除去対象
text = "\nHello, World!\n"
print(text.strip())            # → Hello, World!

# 特定の文字を除去
filename = "***report***"
print(filename.strip("*"))     # → report
````

## 一部分を取り出す（スライス）

文字列の一部だけ取り出したいときは `[開始:終了]` の記法を使います。日付文字列から年月日を取り出すような場面で便利です。

⚠️ **終了位置は「そこまでで止まる」（含まない）** ことに注意してください。

### VBA との対比

| やること | VBA | Python |
|---|---|---|
| 1文字目 | `Mid(s, 1, 1)` | `s[0]` （**0始まり**） |
| 1〜3文字目 | `Mid(s, 1, 3)` | `s[0:3]` または `s[:3]` |
| 後ろから3文字 | `Right(s, 3)` | `s[-3:]` |
| 4文字目以降 | `Mid(s, 4)` | `s[3:]` |

Python は **インデックスが 0 始まり**（先頭が `s[0]`）、`-1` は末尾を表します。VBA が 1 始まりだったので、ここは慣れが必要です。

````python
s = "2025-03-15"

# 個別の文字
print(s[0])     # → 2 （先頭）
print(s[-1])    # → 5 （末尾）

# 範囲 [開始:終了]（終了は含まない）
print(s[0:4])   # → 2025
print(s[5:7])   # → 03
print(s[8:10])  # → 15

# 省略形
print(s[:4])    # → 2025（先頭から4文字）
print(s[5:])    # → 03-15（5文字目以降）
print(s[-2:])   # → 15（末尾2文字）
````

## よくあるエラーと対処

### `f` を付け忘れた

```python
name = "田中"
print("こんにちは {name} さん")     # ❌ → こんにちは {name} さん  (文字列のまま)
print(f"こんにちは {name} さん")    # ⭕ → こんにちは 田中 さん
```

→ 文字列の前の `f` を忘れると、`{name}` がそのまま文字列として表示されます。

### `IndexError: string index out of range`

```python
s = "abc"
print(s[5])    # IndexError
```

→ 文字数を超えるインデックスを指定したときに発生。`len(s)` で長さを確認してからアクセスしましょう。

### `replace()` してるのに元の文字列が変わらない

```python
text = "hello world"
text.replace("world", "Python")
print(text)    # → hello world （変わってない！）
```

→ Python の文字列は **置換しても元の変数は変わらない**（新しい文字列が返ってくるだけ）。結果を **代入し直す** 必要があります：

```python
text = text.replace("world", "Python")
print(text)    # → hello Python
```
