## 関数とは

関数は **「ひとまとまりの処理に名前を付けて、後から何度でも呼び出せるようにしたもの」** です。

VBA で言う `Sub` や `Function` と同じ役割で、Python では **どちらも `def`** で書きます。「戻り値があるか」だけで使い分けます。

### VBA との対比

```vba
' VBA: 戻り値なし (Sub)
Sub Greet(name As String)
    Debug.Print "こんにちは、" & name & "さん"
End Sub

' VBA: 戻り値あり (Function)
Function AddTax(price As Long) As Long
    AddTax = price * 1.1
End Function
```

```python
# Python: どちらも def
def greet(name):
    print(f"こんにちは、{name}さん")

def add_tax(price):
    return price * 1.1
```

ポイント：

- 1行目は `def 関数名(引数):` で、行末は **`:`**（`if` や `for` と同じルール）
- 中身は **インデント** で表す（`End Sub` / `End Function` は不要）
- 値を返したいときは **`return`** を書く（VBA のように `関数名 = 値` ではない）
- 関数名は **小文字 + アンダースコア**（`snake_case`）が Python の慣習

## やってみる: 一番シンプルな関数

まずは引数も戻り値もない、ただ表示するだけの関数を書いてみます。

````python
# 関数の定義
def say_hello():
    print("こんにちは！")
    print("今日もお疲れさまです")

# 関数を呼び出す（カッコを忘れない）
say_hello()

# 何度でも呼べる
say_hello()
say_hello()
````

**ポイント**: 関数を `def` で **定義しただけ** では実行されません。`say_hello()` のように **カッコ付きで呼び出して** 初めて中身が動きます。

> ⚠️ `say_hello` だけ書くと「関数そのもの」を指しているだけで、実行されません。必ず `()` を付ける。

## 引数を受け取る関数

関数に **外から値を渡せる** ようにすると、同じ処理を別のデータに対して再利用できます。渡す値のことを **引数（ひきすう）** と言います。

````python
# name という引数を受け取る
def greet(name):
    print(f"こんにちは、{name}さん")

greet("田中")    # → こんにちは、田中さん
greet("鈴木")    # → こんにちは、鈴木さん

# 引数を2つ取る
def show_sale(item, amount):
    print(f"{item}: ¥{amount:,}")

show_sale("りんご", 150)
show_sale("メロン", 2800)
````

## 戻り値を返す関数 — `return`

関数の中で計算した結果を **「関数の外に持ち出す」** ためのキーワードが `return` です。VBA の `Function` で `関数名 = 値` と書くところを、Python では `return 値` で書きます。

`return` した値は、関数を呼び出した側で **変数に代入したり、`print` したり、別の計算に使ったり** できます。

````python
# 税込み価格を計算して返す関数
def add_tax(price):
    return price * 1.1

# 結果を変数で受け取る
total = add_tax(1000)
print(total)   # → 1100.0

# print の中で直接使ってもいい
print(f"税込: ¥{add_tax(2980):,.0f}")   # → 税込: ¥3,278

# 別の計算にも使える
grand_total = add_tax(1000) + add_tax(2000)
print(grand_total)   # → 3300.0
````

### `return` がない関数 vs ある関数

両者は似ているようで **使い方が全然違う** ので、はじめに整理しておきます。

| | `return` なし（`print` だけ） | `return` あり |
|---|---|---|
| 画面に出る | はい | いいえ（自分で `print` する必要あり）|
| 結果を変数に入れられる | いいえ（`None` が入る）| はい |
| 別の計算に使える | いいえ | はい |
| 用途 | 「表示する」だけのとき | 「計算結果を後で使う」とき |

````python
# 比較してみる
def print_tax(price):
    print(price * 1.1)        # 画面に出すだけ

def return_tax(price):
    return price * 1.1        # 値を返す

# print_tax の戻り値を変数に入れると…
x = print_tax(1000)   # 画面に 1100.0 と出る
print(x)              # → None（何も返していないので）

# return_tax の戻り値は変数に入る
y = return_tax(1000)
print(y)              # → 1100.0
````

## デフォルト引数 と キーワード引数

Python の関数は、**引数にデフォルト値を設定** したり、呼び出すときに **「引数の名前を指定して」渡す** ことができます。VBA の **名前付き引数** (`Greet name:="田中"`) と同じ仕組みです。

### デフォルト引数: 「省略したらこの値を使う」

`def f(税率=0.10):` と書くと、呼び出すとき省略すれば自動的に `0.10` が使われます。

### キーワード引数: 「名前で渡す」

引数が多い関数を呼ぶとき、`f(rate=0.08)` のように名前指定で渡すと、**何の引数か一目で分かる** し、**順番を気にしなくていい** ので読みやすくなります。

````python
# rate にデフォルト値を設定（省略時は 0.10 = 10%）
def add_tax(price, rate=0.10):
    return price * (1 + rate)

# rate を省略 → デフォルトの 10% が使われる
print(add_tax(1000))             # → 1100.0

# rate を渡せば上書きできる
print(add_tax(1000, 0.08))       # → 1080.0（軽減税率）

# キーワード引数で渡すと意図が明確
print(add_tax(1000, rate=0.08))  # → 1080.0

# 引数が多い関数だと特に便利
def make_label(item, price, quantity=1, discount=0):
    final = price * quantity * (1 - discount)
    return f"{item} ×{quantity}個 → ¥{final:,.0f}"

# 順番通りに渡す
print(make_label("りんご", 150, 3))

# キーワードで指定（discount だけ指定したい時に便利）
print(make_label("メロン", 2980, discount=0.2))   # quantity は省略 → デフォルト1
````

## スコープ — 関数の中の変数は外に出ない

関数の中で作った変数は、**その関数の中でしか使えません**。関数を抜けると消えてしまいます。これを **スコープ（変数の有効範囲）** と呼びます。

逆に、関数の **外で作った変数** は関数の中から「読む」ことはできますが、**書き換えはできません**（明示的に `global` 宣言しない限り）。

VBA でも `Sub` の中の変数は基本ローカルなので、感覚はほぼ同じです。

````python
def calculate():
    inner = 100          # 関数の中で作った変数
    print(f"中: {inner}")

calculate()              # → 中: 100

# 関数の外から inner を見ようとするとエラー
# print(inner)   # ← コメントを外すと NameError になる

# 「結果を外で使いたい」なら return で返す
def calculate2():
    inner = 100
    return inner

result = calculate2()
print(f"外: {result}")   # → 外: 100
````

## ちょっと実務寄り: 集計処理を関数化する

1-4 で書いた **「売上リストをランク別に集計する」** 処理を、**関数にまとめて月ごとに再利用** してみます。

毎月同じ集計を手で書き直すのは大変ですが、関数化しておくと **データを渡すだけで何度でも呼べる** ので、本当に楽になります。これが「自動化」の第一歩です。

````python
def summarize_sales(sales):
    """売上リストを大口/中口/小口に分けて集計する"""
    summary = {
        "大口 (2万以上)": {"count": 0, "total": 0},
        "中口 (1万〜2万未満)": {"count": 0, "total": 0},
        "小口 (1万未満)": {"count": 0, "total": 0},
    }
    for s in sales:
        if s >= 20000:
            rank = "大口 (2万以上)"
        elif s >= 10000:
            rank = "中口 (1万〜2万未満)"
        else:
            rank = "小口 (1万未満)"
        summary[rank]["count"] += 1
        summary[rank]["total"] += s
    return summary

def print_summary(title, summary):
    """集計結果を整形して表示する"""
    print(f"=== {title} ===")
    for rank, data in summary.items():
        print(f"{rank:<20} {data['count']}件   合計 ¥{data['total']:,}")
    print()

# 月ごとに呼び出すだけで集計できる
sales_4月 = [12000, 8500, 15000, 9800, 22000, 7600, 35000, 4200, 18000]
sales_5月 = [9000, 25000, 11000, 6500, 30000, 14000]

print_summary("4月の集計", summarize_sales(sales_4月))
print_summary("5月の集計", summarize_sales(sales_5月))
````

**ポイント:**

- `""" ... """` の3連クォートは **docstring**（関数の説明）。関数の1行目に書く慣習で、後から `help(関数名)` で確認できます
- 関数を **「集計する」**「**表示する**」と役割で分けたことで、月が増えても呼ぶ行を増やすだけで済む
- これが第4章の **総集編「月次売上レポート自動生成」** に直結する考え方です

## よくあるエラーと対処

### `def` 行のコロン `:` 忘れ

```python
def greet(name)        # ❌ 行末の : が抜けている
    print(name)
```

→ `def 関数名(引数):` の **行末は必ず `:`**。`if` / `for` と同じルールです。

### 関数を呼ぶときに `()` を付け忘れる

```python
def say_hello():
    print("hi")

say_hello   # ← これだと「関数オブジェクト」を指すだけで、実行されない
```

→ 呼び出すときは **必ず `say_hello()`** とカッコを付ける。

### `return` を忘れて `None` が返る

```python
def add_tax(price):
    price * 1.1   # ← 計算しただけで return していない

result = add_tax(1000)
print(result)   # → None
```

→ 値を外で使いたいときは **必ず `return`** を書く。VBA の `関数名 = 値` のクセで忘れがちな所です。

### `TypeError: f() missing 1 required positional argument: 'x'`

```python
def add(a, b):
    return a + b

add(10)   # ❌ 引数が足りない
```

→ 引数の数が足りない。デフォルト引数 `b=0` を付けるか、呼び出し側で `add(10, 20)` のように渡す。

### `NameError: name 'inner' is not defined`

```python
def calc():
    inner = 100

calc()
print(inner)   # ❌ 関数の中の変数は外で見えない
```

→ 関数の中で作った変数は **外で見えない**（スコープ）。外で使いたいなら `return` で持ち出す。

## まとめ

このノートブックで学んだこと：

- **関数** は `def 関数名(引数):` で定義、**`()` を付けて呼び出す**
- **`return`** で値を外に持ち出す（書かないと `None` が返る）
- **デフォルト引数** `def f(rate=0.10):` で省略時の値を決められる
- **キーワード引数** `f(rate=0.08)` で名前指定して渡せる（読みやすい）
- **スコープ**: 関数内の変数は外に出ない（必要なら `return` で返す）
- **複数の戻り値** は `return a, b` で返せる（受け取りは `x, y = f()`）
- 同じ処理を **関数にまとめる** ことで、データだけ差し替えて何度でも呼べる
