## 3分デモ：Python ではこんなことができる

本格的に学び始める前に、ゴールのイメージを掴んでもらいます。

下のセルは、**サンプル売上データから「月次の推移」と「商品別の売上」をダッシュボード風に1枚にまとめる** までを、一気に行うコードです。

コードの意味はまだ分からなくて大丈夫。次の2点だけ感じ取ってください。

- **Excel グラフで作るとそこそこ手間がかかる体裁を、Python ならコードでさっと出せる**
- **VBA で書いたら何百行になりそうな処理も、これくらいの量で終わる**

```python
# 日本語をグラフに表示するための準備（Colab で最初の1回だけ実行）
!pip install japanize-matplotlib -q
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib  # 日本語フォントを有効にする
import numpy as np

# ─────────────────────────────
# サンプルの売上データを作成
# ─────────────────────────────
np.random.seed(42)
df = pd.DataFrame({
    "日付": pd.date_range("2025-01-01", periods=120),
    "商品": np.random.choice(["商品A", "商品B", "商品C"], 120),
    "売上": np.random.randint(800, 3000, 120),
})

# ─────────────────────────────
# 集計（月別 / 商品別）
# ─────────────────────────────
df["月"] = df["日付"].dt.to_period("M").astype(str)
monthly = df.groupby("月")["売上"].sum()
by_product = df.groupby("商品")["売上"].sum().sort_values()

# ─────────────────────────────
# グラフ描画（2ペインのダッシュボード）
# ─────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("売上ダッシュボード（サンプル）", fontsize=17, fontweight="bold", y=1.02)

# 左: 月別の売上推移（折れ線 + 塗りつぶし）
axes[0].plot(monthly.index, monthly.values,
             marker="o", markersize=10, linewidth=2.5, color="#4C72B0")
axes[0].fill_between(range(len(monthly)), monthly.values, alpha=0.15, color="#4C72B0")
axes[0].set_title("月別の売上推移", fontsize=14, fontweight="bold", pad=10)
axes[0].set_ylabel("売上（円）", fontsize=11)
axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)
axes[0].grid(axis="y", alpha=0.3, linestyle="--")

# 右: 商品別の売上合計（横棒 + 値ラベル）
colors = ["#4C72B0", "#DD8452", "#55A868"]
bars = axes[1].barh(by_product.index, by_product.values,
                    color=colors, edgecolor="white", linewidth=2)
for bar, value in zip(bars, by_product.values):
    axes[1].text(value + max(by_product.values) * 0.01,
                 bar.get_y() + bar.get_height() / 2,
                 f"¥{value:,}",
                 va="center", fontsize=11, fontweight="bold")
axes[1].set_title("商品別の売上合計", fontsize=14, fontweight="bold", pad=10)
axes[1].set_xlim(0, max(by_product.values) * 1.18)
axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)
axes[1].spines["left"].set_visible(False)
axes[1].tick_params(left=False)
axes[1].grid(axis="x", alpha=0.3, linestyle="--")

plt.tight_layout()
plt.show()
```

ダッシュボードが2枚並んで表示されれば、Colab 環境のセットアップは完璧です。

ここで使った 2 つのライブラリが、この講座のメインです。

- **pandas（パンダス）** … Excel/CSV のデータをプログラム上で扱うための道具（第2章で学びます）
- **matplotlib（マットプロットリブ）** … グラフを描くための道具（第3章で学びます）

「グラフの体裁を整える書き方」は、第3章でひとつずつ分解して習得していきます。今はコピペで動けば OK。
