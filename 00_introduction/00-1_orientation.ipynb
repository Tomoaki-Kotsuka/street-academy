## 3分デモ：Python ではこんなことができる

本格的に学び始める前に、ゴールのイメージを掴んでもらいます。

下のセルは、**売上明細のサンプルデータを作って → 商品ごとに集計して → グラフを描く** までを、たった10行程度で実現するコードです。

コードの意味はまだ分からなくて大丈夫です。「**VBA で書いたら何百行になりそうな処理が、これだけで終わるんだ**」という感覚だけ味わってください。

```python
# 日本語をグラフに表示するための準備（Colab で最初の1回だけ実行）
!pip install japanize-matplotlib -q
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib  # 日本語フォントを有効にする
import numpy as np

# ─────────────────────────────
# サンプルの売上データを作成
# ─────────────────────────────
np.random.seed(42)
df = pd.DataFrame({
    "日付": pd.date_range("2025-01-01", periods=60),
    "商品": np.random.choice(["商品A", "商品B", "商品C"], 60),
    "売上": np.random.randint(800, 3000, 60),
})

# ─────────────────────────────
# 商品ごとに売上を集計
# ─────────────────────────────
summary = df.groupby("商品")["売上"].sum()
print(summary)

# ─────────────────────────────
# グラフ化
# ─────────────────────────────
summary.plot(kind="bar")
plt.title("商品別の売上合計")
plt.ylabel("売上（円）")
plt.show()
```

棒グラフが表示されれば、Colab 環境のセットアップは完璧です。

ここで使った 2 つのライブラリが、この講座のメインです。

- **pandas（パンダス）** … Excel/CSV のデータをプログラム上で扱うための道具（第2章で学びます）
- **matplotlib（マットプロットリブ）** … グラフを描くための道具（第3章で学びます）
