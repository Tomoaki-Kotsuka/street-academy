第 4 章は **「自分の PC のフォルダにあるファイルを Python が直接読む」** という、実務に直結する状態で動かす章です。Colab のままでも進められますが、**ローカル環境** にすると次のメリットがあります:

- **47 個の Excel ファイルがあっても、毎回アップロード/ダウンロードしなくていい**

## このノートのゴール

- **uv** を使って、自分の PC に **Python と必要なライブラリ** をワンセットで入れる

- 第 4 章の作業フォルダ構成を整える

## なぜ uv？

**Python のローカル構築は、これまで一番のハードル** でした。

- 公式サイトから Python をダウンロード → インストーラ実行 → PATH 通す
- `pip install` でライブラリを入れる → バージョンが衝突する
- 別のプロジェクトとライブラリが混ざらないように `venv` で仮想環境を作る
- …これで挫折する人が大半でした。

**uv は、これを全部 1 つにまとめてくれる新しいツール** です。

- **Python 本体のインストールも uv がやる** → 公式サイトに行かなくていい
- **依存ライブラリは 1 コマンドで追加** → `uv add pandas` だけ
- **仮想環境は uv が自動で作る** → 意識しなくていい
- **爆速** — Rust 製で、pip の数十倍速く動く
- **無料・OSS** — Anaconda のような商用利用ライセンスの心配なし

## ステップ 1: uv をインストール

ターミナル（**Windows なら PowerShell、Mac なら ターミナル.app**）を開いて、以下を 1 行コピペして実行してください。

### Windows

```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

### Mac

```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

> ⚠️ **インストール後、ターミナルを一度閉じて開き直してください**（PATH を反映するため）。

### 確認

新しいターミナルで:

```bash
uv --version
```

→ `uv 0.x.x` のようにバージョンが表示されれば成功です。

## ステップ 2: プロジェクトフォルダを作る

好きな場所にプロジェクト用のフォルダを作ります。ここでは **デスクトップに `python-data-basics-local`** を作る例で進めます。

### Windows (PowerShell)

```powershell
cd $HOME\Desktop
mkdir python-data-basics-local
cd python-data-basics-local
```

### Mac (ターミナル)

```bash
cd ~/Desktop
mkdir python-data-basics-local
cd python-data-basics-local
```

### プロジェクトを初期化

Windows / Mac 共通:

```bash
uv init
```

→ フォルダの中に `pyproject.toml` `README.md` `main.py` などが自動で作られます。

## ステップ 3: 必要なライブラリをまとめて追加

第 4 章で使うライブラリを 1 コマンドで入れます。

```bash
uv add pandas openpyxl matplotlib japanize-matplotlib jupyterlab
```

それぞれの役割:

| ライブラリ | 役割 |
|---|---|
| `pandas` | 表データを扱う（2 章で学んだやつ） |
| `openpyxl` | Excel ファイルを読み書きする |
| `matplotlib` | グラフを描く（3 章で学んだやつ） |
| `japanize-matplotlib` | matplotlib のグラフで日本語が出る |
| `jupyterlab` | `.ipynb` をブラウザで開いて実行する |

> ⚙️ 初回は依存ライブラリも合わせて 100 個近くダウンロードされますが、uv は爆速なので **数秒で終わります**。

## ステップ 4: サンプルデータとノートを配置

GitHub のリポジトリから、**第 4 章のサンプルデータ** と **ノートブック一式** を `python-data-basics-local/` に持ってきます。

`python-data-basics-local/` の中がこうなっていれば OK:

## ステップ 5: Jupyter Lab を起動して `.ipynb` を開く

`python-data-basics-local/` フォルダの中で:

```bash
uv run jupyter lab
```

**何が起きるか:**

- ターミナルに `http://localhost:8888/lab?token=...` のような URL が表示される
- **ブラウザが自動で開く**（開かない場合はその URL をブラウザに貼る）
- 左サイドバーにフォルダツリーが見える → `04_capstone/04-1_overview.ipynb` などをダブルクリックで開ける

### 終了するとき

- ブラウザのタブは閉じて OK
- ターミナルに戻って **`Ctrl + C` を 2 回** 押して Jupyter Lab を止める

### 次回以降

また学習を再開するときは、ターミナルで `python-data-basics-local/` に `cd` してから `uv run jupyter lab` を打つだけ。**uv が裏で自動的に Python と仮想環境を起動してくれます**。

## Colab との違い — 何を書き換える？

**書き換えるのは 4-1 以降の最初のセルだけ** です。

### Colab 版（これまでの章のスタイル）

```python
from google.colab import drive
drive.mount('/content/drive')
BASE = "/content/drive/MyDrive/python-data-basics"
```

### ローカル版

```python
# ローカルでは drive.mount は不要。今いるフォルダを基準にする
BASE = "."   # python-data-basics-local/ の中で起動している前提
```

→ 以後の `f"{BASE}/data/capstone/..."` のようなパス組み立てコードは **Colab とローカルで完全に同じ** ものが動きます。これが、Colab/ローカルどちらでも進められる仕掛けです。

## よくあるエラー

### 1. `uv: command not found`
→ インストール直後でターミナルを開き直していない。一度閉じて開き直す。それでも出ない場合は PATH が通っていないので、ステップ 1 の手順を再実行。

### 2. `uv init` で「This directory is not empty」と怒られる
→ 既に何かファイルがある状態で初期化しようとしている。`uv init --bare` を使うか、空のフォルダを別に作ってやり直す。

### 3. `uv add` で「Failed to fetch」エラー
→ ネットワーク（特に会社 PC のプロキシ）が原因のことが多い。会社 PC で進める場合は、情シス部門に確認するか、自宅 PC で進める。

### 4. `uv run jupyter lab` でブラウザが開かない
→ ターミナルに表示された `http://localhost:8888/lab?token=...` を**手動でブラウザに貼り付ける**。Token 込みの URL でないと弾かれます。

### 5. Jupyter Lab で `.ipynb` を開いてもセルが実行できない
→ 右上のカーネル表示が「No Kernel」になっていないか確認。**Kernel → Change Kernel → Python 3 (ipykernel)** を選び直す。
